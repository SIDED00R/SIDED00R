```python
from sklearn import svm
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import random

d=datasets.load_iris()
s=svm.SVC(gamma=0.1,C=10)
idx = list(range(len(d.data)))
random.shuffle(idx)
X_train = [d.data[i] for i in idx[:20]]
y_train = [d.target[i] for i in idx[:20]]

for i in range(0,len(X_train)):
  print(X_train[i])

print("새로운 샘플의 부류는 ",y_train )
s.fit(X_train,y_train)

rannum=random.randrange(1,21)
print("new random sample index: ",rannum)

ransel=random.randrange(21,151)
X_train[rannum-1]=d.data[ransel-1]
y_train[rannum-1]=d.target[ransel-1]

for i in range(0,len(X_train)):
  print(X_train[i])
print("변형된 샘플의 부류는 ",y_train )

res=s.predict(X_train)
correct=[i for i in range(len(res)) if res[i]==y_train[i]]
accuracy=len(correct)/len(res)
print(accuracy*100)
```

```python
from sklearn import datasets
from sklearn import svm
from sklearn.model_selection import cross_val_score
import numpy as np

for i in range(5,11):
  digit=datasets.load_digits()
  s=svm.SVC(gamma=0.001)
  accuracies=cross_val_score(s,digit.data,digit.target,cv=i)
  print(i,"겹 교차검증")
  print(accuracies)
  print("정확률(평균)=%0.3f, 표준편차=%0.3f"%(accuracies.mean()*100,accuracies.std()))
```
