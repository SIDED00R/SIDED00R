```python
import torch
import numpy as np
import matplotlib.pyplot as plt

x_data=[1,2,3]
y_data=[2,4,6]
w=1
def forward(x):
  return x*w
def loss(x,y):
  y_pred=forward(x)
  return (y_pred-y)*(y_pred-y)
def gradient(x,y):
  return 2*x*(x*w-y)
for epoch in range(100):
  l=0
  for x_val,y_val in zip(x_data,y_data):
    grad=gradient(x_val,y_val)
    w=w-0.01*grad
    l=l+loss(x_val,y_val)
  mse=l/3
  print("progress:", epoch, "w=",w,"mse=", mse)
print("test performance (after training) ", 4, forward(4))
```
