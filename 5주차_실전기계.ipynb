```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=torch.ones(3,2)
print(x)
x=torch.zeros(3,2)
print(x)
```

```python
torch.manual_seed(2)
x=torch.rand(3,2)
print(x)
x=torch.randn(3,3)
print(x)
```

```python
x=torch.tensor([[1,2],
               [3,4],
               [5,6]])
print(x[:,1])
print(x[0,:])
y=x[1,1]
print(y)
```

```python
x=torch.tensor([[1,2],
               [3,4],
               [5,6]])
y=x.view(2,3)
print(y)
y=x.view(6,-1) #-1은 자동으로 숫자 채워줌
print(y)
```

```python
x=torch.ones([3,2])
y=torch.ones([3,2])
z=x+y
print(z)
z=torch.add(x,y)
print(z)
z=x-y
print(z)
z=torch.sub(x,y)
print(z)

#y=y+x -> y+=x
y.add_(x) #y+=x
print(y)
```

```python
x=torch.linspace(0,1,steps=5)
print(x)
x_np=x.numpy()
print(type(x),type(x_np))
```

```python
a=np.random.randn(5)
a_pt=torch.from_numpy(a)
print(a)
print(type(a),type(a_pt))
```

```python
x=torch.ones([3,2],requires_grad=True)
print(x)
```

```python
y=x+5
print(y)

z=y*y+1
print(z)
t=torch.sum(z)
print(t)
```

```python
t.backward()
print(x.grad)
```
