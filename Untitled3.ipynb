```python
from sklearn import datasets
from sklearn.model_selection import  train_test_split
import numpy as np
from sklearn.linear_model import Perceptron

digit=datasets.load_digits()
x_tr,x_te,y_tr,y_te=train_test_split(digit.data,digit.target,train_size=0.6)

p=Perceptron(max_iter=100, eta0=0.001, verbose=1)
p.fit(x_tr,y_tr)
res=p.predict(x_te)

conf=np.zeros((10,10))
for i in range(len(res)):
  conf[res[i]][y_te[i]]+=1
print(conf)

no_correct=0
for i in range(10):
  no_correct+=conf[i][i]
accuracy=no_correct/len(res)
print(accuracy*100)
```
